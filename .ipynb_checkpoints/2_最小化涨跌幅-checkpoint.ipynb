```python
import numpy as np
import pandas as pd
```

## 读取数据

```python
# 成分股数据
df=pd.read_excel("df_all.xlsx",index_col=0)
df.head()
```

```python
# 沪深300指数数据
df300=pd.read_excel("df_300_w.xlsx",index_col=0)
df300.head()
```

```python
# 根据权重从大到小选择前20支股票
df20=df300.loc[:19,:]
df20.head()
```

## 变成时间序列数据
每一列为一个股票在3年的涨跌幅数据

```python
df1=pd.pivot_table(df,index=["trade_date","ts_code"],values=["pct_chg"])
df1.head()
```

找一天算一下涨跌幅

```python
# 300支股票20241211涨跌幅
df_i=df1.query('trade_date == [20211210]')
df_i=df_i.reset_index()
df_i.head()
```

```python
# 连表
df_i_w=pd.merge(df_i,df300[['权重(%)weight','ts_code']],how='inner',on='ts_code')
df_i_w.head()
```

```python
# # len(df_i_w)
# len(df_i)
```

```python
# 计算
chg=np.sum(np.array(df_i_w['pct_chg'].tolist())/100*np.array(df_i_w['权重(%)weight'].tolist())/100)*100
chg
```

## 找到指数权重前20支股票的近3年每天的涨跌幅时序数据

```python
ts_code_20=df20['ts_code'].tolist()
df.head()
```

```python
dff=pd.DataFrame(columns=list(df.columns))
for i in df.index:
    if df.loc[i,'ts_code'] in ts_code_20:
        dff=pd.concat([dff,df.loc[i,:]])
dff.head()
```

## 开始优化
现在有数据：<br />
1. 沪深300指数数据：df300
2. 沪深300指数近3年每天的数据：df300_daily
3. 成分股近3年每天的数据：df
4. 成分股近3年涨跌幅时序数据：df1
5. 指数权重前20支股票的代码与权重值：df20
<br />

目标是最小化绝对误差，找到一组不超过20支股票买入的权重w

```python
import mfo
```

```python
# 方法：飞蛾扑火优化算法
lb=np.array([0]*20) # 下界
ub=np.array([1]*20) # 上界
max_i=100
dim=20
pop=100
chgi=df1
chg300=df300_daily[['日期Date','涨跌幅(%)Change(%)']]
p=10 # 惩罚因子
index2codes=ts_code_20
```

```python
mymfo=mfo.MFO(pop,max_i,dim,ub,lb,chgi,chg300,index2codes,p)
```

```python
mymfo.run()
```
