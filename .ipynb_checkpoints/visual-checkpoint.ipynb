```python
df1=pd.read_excel("../000300closeweight.xlsx")
df1=df1.loc[:19,:]
df1.head()
```

```python
df1["地区"]=df1["交易所Exchange"].map({
    "上海证券交易所":'SH',
    "深圳证券交易所":'SZ'
})
df1.head()
```

```python
ts_code_top_20=df1["ts_code"].tolist()
ts_code_top_20
```

```python
ts.set_token('35f337e592e9520154d2a0827c491f75ccc87b723d2bec02cea6b78b')
pro = ts.pro_api()
cols=['ts_code','trade_date','open','high','low','close','pre_close','change','pct_chg','vol','amount']
df_all=pd.DataFrame(columns=cols)
df_all.head()
```

```python
for cd in ts_code_top_20:
    df_i = pro.daily(ts_code=cd, start_date='20240913', end_date='20241211')
    df_all=pd.concat([df_all, df_i])
    df_i.to_excel("../ts_top_20_20240913_20241211.xlsx",sheet_name=cd)
```
