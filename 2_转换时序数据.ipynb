## 变成时间序列数据
每一列为一个股票在3年的涨跌幅数据

```python
df1=pd.pivot_table(df,index=["trade_date","ts_code"],values=["pct_chg"])
df1.head()
```

找一天算一下涨跌幅

```python
# 300支股票20241211涨跌幅
df_i=df1.query('trade_date == [20211210]')
df_i=df_i.reset_index()
df_i.head()
```

```python
# 连表
df_i_w=pd.merge(df_i,df300[['权重(%)weight','ts_code']],how='inner',on='ts_code')
df_i_w.head()
```

```python
# # len(df_i_w)
# len(df_i)
```

```python
# 计算
chg=np.sum(np.array(df_i_w['pct_chg'].tolist())/100*np.array(df_i_w['权重(%)weight'].tolist())/100)*100
chg
```
